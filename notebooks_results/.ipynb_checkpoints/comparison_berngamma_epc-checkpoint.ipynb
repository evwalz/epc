# Comparison between EPC 15 and BG

BG (Bernoulli-Gamma) distribution paramters are obtained via scripts in folder [script_bernoulligamma](script_bernoulligamma). 

Compute CRPS of BG and EPC 15 on the years 2018 and 2019 with *bg_crps_year_18_19.py* and *epc_crps_year_18_19.py* in [scripts_crps](scripts_crps) 

```python
import xarray as xr
import numpy as np
import pandas as pd
from scipy.stats import rankdata
import matplotlib.pyplot as plt
 
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import cartopy.crs as ccrs
```

```python
data_dir = '/Users/eva/Documents/Work/data/'
```

```python
bern_crps = xr.open_dataset(data_dir + "bern_crps_1819.nc")
epc_crps = xr.open_dataset(data_dir + "epc_crps_1819.nc4")
land_sea = xr.open_dataset(data_dir + "GPM_IMERG_LandSeaMask.2.nc4")
land_sea_new = land_sea.assign_coords(lon = epc_crps.lon.values)
```

```python
# functions
def teststatistic(epc1, epc2, days):
    Smean_F = epc1.mean(dim="time")
    Smean_G = epc2.mean(dim="time")
    difference = np.subtract(Smean_F,Smean_G)
    factor = np.sqrt(days)
    variance = np.square(np.subtract(epc1,epc2)).mean(dim="time")
    sigma = np.sqrt(variance)
    teststatistic = factor*np.divide(difference, sigma)
    return(teststatistic)
```

### TODM equivalence test

```python
# TODM between BG and EPC 15
testval = teststatistic(epc_crps, bern_crps, 2*365)
testval = testval.precipitationCal.values

variance = (np.square(epc_crps - bern_crps)).mean('time')
sigma = np.sqrt(variance).precipitationCal.values
```

```python
from scipy.stats import norm

thetas = np.arange(0.005, 0.151, 0.001)
alpha = 0.05
days = 2 * 365
percent_test = []
for theta in thetas:
    thetacorrection = (theta*np.sqrt(days))/sigma
    test_l = testval + thetacorrection
    test_u = thetacorrection - testval
    dimspace = testval.shape[0]*testval.shape[1]
    ds = dimspace / 100
    compare = np.arange(1, dimspace+1)
 
    pval_l = 1-norm.cdf(test_l.flatten())
    pval_u = 1-norm.cdf(test_u.flatten())
    out_l = np.where(np.sort(pval_l) <= compare*alpha/dimspace, 1, 0)
    out_u = np.where(np.sort(pval_u) <= compare*alpha/dimspace, 1, 0)
    indx1 = np.argsort(pval_l)
    indx2 = np.argsort(pval_u)
    a = np.column_stack((out_l,indx1))
    a = a[a[:,1].argsort()]
    a2 = np.where(a[:,0] == 1)
    b = np.column_stack((out_u,indx2))
    b = b[b[:,1].argsort()]
    b2 = np.where(a[:,0] == 1)
    percent_test.append(np.round(len(np.unique(np.concatenate([a2,b2], axis=1)))/ds,2))
```

```python
import matplotlib.pyplot as plt
plt.plot(np.arange(0.005, 0.151, 0.001), np.array(percent_test))#, color = '#08519c')
plt.ylabel('Significant grid points [%]')
plt.xlabel("\u03b8 [mm]")
plt.xticks(np.arange(0,0.151, 0.03))
plt.grid()
```

```python
# TODM between BG and EPC 15 over land
sel_lat = land_sea.lat.values[500:1300]
land_sea_sel = land_sea.sel(lat = sel_lat) 
ls_bin = land_sea_sel.landseamask.values
ls_indx_land = np.where(ls_bin > 50)

ls_flat = ls_bin.flatten()
indx_ocean = np.where(ls_flat > 50)

# flatten testval and remove sea indices
testval_flat = testval.flatten()
testval_land = np.delete(testval_flat, indx_ocean)
sigma_flat = sigma.flatten()
sigma_land = np.delete(sigma_flat, indx_ocean)
```

```python
from scipy.stats import norm

thetas = np.arange(0.001, 0.051, 0.001)
alpha = 0.05
days = 2 * 365
percent_test = []
for theta in thetas:
    thetacorrection = (theta*np.sqrt(days))/sigma_land
    test_l = testval_land + thetacorrection
    test_u = thetacorrection - testval_land
    dimspace = len(testval_land)
    ds = dimspace / 100
    compare = np.arange(1, dimspace+1)
 
    pval_l = 1-norm.cdf(test_l)
    pval_u = 1-norm.cdf(test_u)
    out_l = np.where(np.sort(pval_l) <= compare*alpha/dimspace, 1, 0)
    out_u = np.where(np.sort(pval_u) <= compare*alpha/dimspace, 1, 0)
    indx1 = np.argsort(pval_l)
    indx2 = np.argsort(pval_u)
    a = np.column_stack((out_l,indx1))
    a = a[a[:,1].argsort()]
    a2 = np.where(a[:,0] == 1)
    b = np.column_stack((out_u,indx2))
    b = b[b[:,1].argsort()]
    b2 = np.where(a[:,0] == 1)
    percent_test.append(np.round(len(np.unique(np.concatenate([a2,b2], axis=1)))/ds,2))
```

```python
import matplotlib.pyplot as plt

plt.plot(thetas, np.array(percent_test))
plt.ylabel('Significant grid points [%]')
plt.xlabel("\u03b8 [mm]")
plt.xticks(np.arange(0,0.051, 0.01))
plt.grid()
```

### mean CRPS for specific regions

```python
# check values of specific areas:

def mean_crps_region(lat_lower, lat_upper, lon_lower, lon_upper):
    np_lon = epc_crps.lon.values
    np_lat = epc_crps.lat.values
    np_lat = np_lat[np_lat >= lat_lower]
    np_lat = np_lat[np_lat <= lat_upper]
    np_lon = np_lon[np_lon >= lon_lower]
    np_lon = np_lon[np_lon <= lon_upper]
 
    crps_lat = e1_mean.sel(lat = np_lat)
    crps_lon = crps_lat.sel(lon = np_lon)

    crps_epc = crps_lon.precipitationCal.values
 
    crps_lat_b = b1_mean.sel(lat = np_lat)
    crps_lon_b = crps_lat_b.sel(lon = np_lon)

    crps_bb = crps_lon_b.precipitationCal.values
 
    ls_lat = np.array(land_sea_new.lat.values)
    ls_lat = ls_lat[ls_lat >= lat_lower]
    ls_lat = ls_lat[ls_lat <=  lat_upper]
     
    ls_lon = np.array(land_sea_new.lon.values)
    ls_lon = ls_lon[ls_lon >= lon_lower]
    ls_lon = ls_lon[ls_lon <= lon_upper]
 
    ls_imerg = land_sea_new.sel(lat = ls_lat)
    ls_imerg2 = ls_imerg.sel(lon = ls_lon)
    ls_bin = ls_imerg2.landseamask.values
    ls_tresh_indx = np.where(ls_bin > 50)
    ls_tresh_indx_l = np.where(ls_bin <= 50)
 
    print('CRPS of EPC over land:', np.round(np.mean(crps_epc[ls_tresh_indx_l]),2))
    print('CRPS of BG over land:', np.round(np.mean(crps_bb[ls_tresh_indx_l]),2))

    print('CRPS of EPC over all grid points:', np.round(np.mean(crps_epc),2))
    print('CRPS of BG over all grid points:', np.round(np.mean(crps_bb),2))
```

```python
# central africa
lat_lower = -15
lat_upper = 5
lon_lower = 12.5
lon_upper = 35.5
mean_crps_region(lat_lower, lat_upper, lon_lower, lon_upper)
```
