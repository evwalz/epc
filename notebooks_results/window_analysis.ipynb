# EPC based on different window sizes 

CRPS values for different EPC forecasts are compared. Code reproduces plots and table from the paper.

To obtain CRPS values for different EPC forecasts run the script 'crps19_test.py' on Imerg dataset first. 

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read in CRPS values for different EPC forecasts. 

years0 = xr.open_mfdataset('crps_0_2*.nc4', concat_dim='year', combine='nested')
years0 = years0.assign_coords(year = np.arange(2001, 2020))

years2 = xr.open_mfdataset('crps_2_*.nc4', concat_dim='year', combine='nested')
years2 = years2.assign_coords(year = np.arange(2001, 2020))

years5 = xr.open_mfdataset('crps_5_*.nc4', concat_dim='year', combine='nested')
years5 = years5.assign_coords(year = np.arange(2001, 2020))

years10 = xr.open_mfdataset('crps_10_*.nc4', concat_dim='year', combine='nested')
years10 = years10.assign_coords(year = np.arange(2001, 2020))

years15 = xr.open_mfdataset('crps_15_*.nc4', concat_dim='year', combine='nested')
years15 = years15.assign_coords(year = np.arange(2001, 2020))

years20 = xr.open_mfdataset('crps_20_*.nc4', concat_dim='year', combine='nested')
years20 = years20.assign_coords(year = np.arange(2001, 2020))
```

```python
lonlat0_mean = years0.mean(dim=['lon', 'lat']).precipitationCal.values
lonlat2_mean = years2.mean(dim=['lon', 'lat']).precipitationCal.values
lonlat5_mean = years5.mean(dim=['lon', 'lat']).precipitationCal.values
lonlat10_mean = years10.mean(dim=['lon', 'lat']).precipitationCal.values
lonlat15_mean = years15.mean(dim=['lon', 'lat']).precipitationCal.values
lonlat20_mean = years20.mean(dim=['lon', 'lat']).precipitationCal.values
```

```python
col1 = '#1f77b4'
col2 = '#00BE67'
col3 = "#ff7f0e"
col4 = '#C77CFF'
col5 = '#d62728'
col6 = '#8c564b'

fig, ax = plt.subplots(figsize=(10, 5))
x = np.arange(19)
plt.plot(x,lonlat0_mean,label='EPC 0', color=col1)
plt.plot(x,lonlat2_mean,label='EPC 2',color=col2)
plt.plot(x,lonlat5_mean,label='EPC 5',color=col3)
plt.plot(x,lonlat10_mean,label='EPC 10',color=col4)
plt.plot(x,lonlat15_mean,label='EPC 15',color=col5)
plt.plot(x,lonlat20_mean,label='EPC 20',color=col6)
plt.ylabel('CRPS [mm]')
plt.xlabel('Year')

my_xticks = ['2001','2004','2007' ,'2010','2013', '2016','2019']
y = [0,3,6,9,12,15,18]
plt.xticks(y, my_xticks)
plt.yticks(np.arange(2.5,2.94,0.05))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#adds major gridlines
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
ax.legend()

#plt.savefig('test4.pdf')
```

```python
print('Mean CRPS for EPC 0: ',np.round(np.mean(lonlat0_mean),2))
print('Mean CRPS for EPC 2: ',np.round(np.mean(lonlat2_mean),2))
print('Mean CRPS for EPC 5: ', np.round(np.mean(lonlat5_mean),2))
print('Mean CRPS for EPC 10: ', np.round(np.mean(lonlat10_mean),2))
print('Mean CRPS for EPC 15: ', np.round(np.mean(lonlat15_mean),2))
print('Mean CRPS for EPC 20: ', np.round(np.mean(lonlat20_mean),2))
```
